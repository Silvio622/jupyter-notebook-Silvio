#### Is changing a good idea?

***

```python
# create a new variable (header) in the Dataframe
df['win'] = True
df
```

```python
#Compariosn of 2 coulmns in the dataframe
df['car'] == df['pick']
```

```python
#Comparison of 2 coulmns in the dataframe and with another columns use the and sign "&"
(df['car'] == df['pick']) & df['stay'] 
```

```python
#Comparison of 2 columns in the dataframe and with another columns use the and sign &"
# in the second argument or input the tilda is used as "False"
((df['car'] == df['pick']) & df['stay']) | ((df['car'] != df['pick']) & ~df['stay'])# ~ tilda is used as['stay']=False
df
```

```python
# show the new dataframe with the win results insert the formula to the win column
df['win'] = ((df['car'] == df['pick']) & df['stay']) | ((df['car'] != df['pick']) & ~df['stay'])# ~ tilda is used as['stay']=False
df
```

```python
# create a catplot how many poeple won
sns.catplot(x="win", kind="count", palette=None, data=df)
```

```python
# half of the people won
df['win'].sum()
```

```python
# bar chart of winning
sns.catplot(x="stay", kind="count", hue='win', data=df)# hue is a different name for color
```
